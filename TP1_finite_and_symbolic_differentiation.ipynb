This code demonstrates two different approaches to differentiation: finite difference approximation and symbolic differentiation. Finite difference approximation is a numerical method that approximates the derivative by evaluating the function at nearby points. Symbolic differentiation, on the other hand, uses mathematical rules to find the exact derivative of a function.

By comparing the results of both methods, you can see how the finite difference approximation becomes more accurate as the step size `h` decreases, while the symbolic differentiation always gives the exact derivative.

This cell serves as an exercise to understand and compare these two approaches to differentiation. You can modify the function `f(x)` and experiment with different step sizes to see how the results change.

```python
import numpy as np
import sympy as sp

import matplotlib.pyplot as plt
%matplotlib inline

# This function is often used in statistics and probability theory
def finite_derivative(f, a, method='central', h=0.01):
    if method == 'central':
        return (f(a + h) - f(a - h)) / (2 * h)
    elif method == 'forward':
        return (f(a + h) - f(a)) / h
    elif method == 'backward':
        return (f(a) - f(a - h)) / h
    else:
        raise ValueError("Method must be 'central', 'forward' or 'backward'.")


if __name__ == '__main__':

    x = sp.symbols('x')
    
    # Define some functions
    f0_expr = 1/(1 + sp.exp(-(x))) - 1.0
    f1_expr = sp.exp(x)
    f2_expr = sp.log(x + 1)
    f3_expr = sp.sin(x)
    f4_expr = 2*x**3 + 3*x**2 - 5*x + 1
    f5_expr = sp.sinh(x)
    f8_expr = x / (x + 1)
    f9_expr = sp.asin(x)
    f10_expr = sp.exp(-x**2)

    # Composition of f1 and f2: (f1 ∘ f2)(x) = f1(f2(x))
    composition_f1_f2  = f1_expr.subs(x, f2_expr)
    composition_f3_f4  = f3_expr.subs(x, f4_expr)
    composition_f9_f10 = f9_expr.subs(x, f10_expr)

    expressions = [f0_expr, f1_expr, 
                   f2_expr, f3_expr, 
                   f4_expr, f5_expr, 
                   f8_expr, f9_expr, f10_expr, 
                   composition_f1_f2, composition_f3_f4, 
                   composition_f9_f10]

    for expr in expressions:

        # Evaluate differentiation

        # FIXME: Convert the sympy expression to a numpy function
        f_func = sp.lambdify(x, expr, 'numpy')

        # FIXME: Compute the derivative of the function
        f_prime_func = sp.lambdify(x, sp.diff(expr, x), 'numpy')

        lst_h = []
        lst_error = []
        nbrange = 100

        ### Numerical differentiation with finite differences with h in [10^-1, 10^-2, ..., 10^-n]
        for h in 10.0**np.arange(-1, -nbrange, -1):
            num_deriv = finite_derivative(f_func, 1, method='central', h=h)
            lst_h.append(h)
            lst_error.append(np.abs(num_deriv - f_prime_func(1)))
            print(f'h={h}, numerical limit={num_deriv}, error (symbolic)={np.abs(num_deriv - f_prime_func(1))}')

        plt.figure()
        plt.loglog(lst_h, lst_error, 'o-', label='Error')
        plt.xlabel('h')
        plt.ylabel('Error')
        plt.title(f'Error in numerical differentiation of {expr}')
        plt.grid()
        plt.show()
```

### Logistic Regression finite differentiation

This example code is an implementation of logistic regression using gradient descent to fit a logistic curve to a given dataset. Here's a step-by-step explanation of what the code does:

1. Importing Libraries: The code begins by importing the necessary libraries. `numpy` is imported as `np` to perform numerical computations, and `matplotlib.pyplot` is imported as `plt` to visualize the data and the fitted logistic curve.

2. Generating Dataset: A simple random dataset is generated using `np.linspace` to create evenly spaced values between -5 and 5. The logistic function is applied to these values, and some random noise is added using `np.random.normal`. The `x_data` represents the input values, and `y_data` represents the corresponding output values.

3. Defining the Logistic Function: The logistic function is defined as `logistic_func(a, b, x)`, which takes in parameters `a`, `b`, and input `x`. It calculates the logistic function using the formula `1 / (1 + np.exp(-(a * x + b)))` and returns the result.

4. Defining the Objective Function: The objective function is defined as `objective_func(a, b)`, which takes in parameters `a` and `b`. It calculates the sum of squared differences between the predicted values obtained from the logistic function and the actual `y_data` values. The formula used is `(logistic_func(a, b, x_data) - y_data)**2`, and the result is returned.

5. Initializing Parameters: The initial values for the parameters `a` and `b` are set as `a_init = 1` and `b_init = 0`. These values can be experimented with to see how they affect the fitting process.

6. Defining the Step Size: The step size `h` is set as `1e-5`. This value is used in the central difference method to compute the gradient.

7. Computing the Gradient: The `numerical_gradient(a, b)` function is defined to compute the gradient of the objective function with respect to the parameters `a` and `b`. The central difference method is used to approximate the gradient by calculating the difference in the objective function values for small perturbations in `a` and `b`. The gradients `grad_a` and `grad_b` are computed using the formulas provided in the code.

8. Performing Gradient Descent: Gradient descent is performed to optimize the parameters `a` and `b`. The learning rate is set as `learning_rate = 0.01`, and the number of iterations is set as `num_iterations = 1000`. In each iteration, the gradients `grad_a` and `grad_b` are computed using the `numerical_gradient` function. The parameters `a_val` and `b_val` are updated by subtracting the product of the learning rate and the gradients. This process is repeated for the specified number of iterations.

9. Obtaining the Optimal Parameters: After the gradient descent process, the optimal values of `a` and `b` are obtained as `a_opt` and `b_opt`, respectively.

10. Plotting the Data and Fitted Logistic Curve: Finally, the data points are plotted using `plt.scatter`, and the fitted logistic curve is plotted using `plt.plot`. The `x_fit` values are generated using `np.linspace` to create a smooth curve, and the corresponding `y_fit` values are obtained by applying the logistic function with the optimal parameters. The data points and the fitted logistic curve are displayed on the plot using `plt.legend` and `plt.show`.

This code demonstrates how logistic regression can be implemented using gradient descent to fit a logistic curve to a given dataset. The resulting curve can be used for classification tasks or to model the relationship between input and output variables.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Generate a simple random dataset
np.random.seed(42)
x_data = np.linspace(-5, 5, num=20)
y_data = 1 / (1 + np.exp(-x_data)) + np.random.normal(scale=0.1, size=len(x_data))

# Define the logistic function
def logistic_func(a, b, x):
    # FIXME
    return 1 / (1 + np.exp(-(a * x + b)))

# Define the objective function
def objective_func(a, b):
    # FIXME 
    return np.sum((logistic_func(a, b, x_data) - y_data)**2)

# Initialize parameters (you can experiment with different initial values)
a_init, b_init = 1, 0

# Define the step size for finite differences
h = 1e-5

# Compute the gradient using central difference method
def numerical_gradient(a, b):
    # FIXME
    grad_a = (objective_func(a + h, b) - objective_func(a - h, b)) / (2 * h)
    grad_b = (objective_func(a, b + h) - objective_func(a, b - h)) / (2 * h)
    return grad_a, grad_b

# Perform gradient descent
learning_rate = 0.01
num_iterations = 1000
a_val, b_val = a_init, b_init

for _ in range(num_iterations):
    grad_a, grad_b = numerical_gradient(a_val, b_val)
    a_val -= learning_rate * grad_a
    b_val -= learning_rate * grad_b

a_opt, b_opt = a_val, b_val

# Plot the data and the fitted logistic curve
plt.scatter(x_data, y_data, label='Data')
x_fit = np.linspace(-5, 5, num=100)
y_fit = logistic_func(a_opt, b_opt, x_fit)
plt.plot(x_fit, y_fit, label='Fitted Logistic Curve', color='red')
plt.legend()
plt.show()
```

Overall, this code demonstrates how to generate a random dataset, define a logistic function symbolically, fit the logistic curve to the data using Sympy's nonlinsolve, and visualize the results using Matplotlib.

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Generate a simple random dataset
np.random.seed(42)
x_data = np.linspace(-5, 5, num=20)
y_data = 1 / (1 + np.exp(-x_data)) + np.random.normal(scale=0.1, size=len(x_data))

# Define the logistic function
x = sp.symbols('x')
a, b = sp.symbols('a, b')
# FIXME: logistic_func = ...
logistic_func = 1 / (1 + sp.exp(-(a * x + b)))

# Fit the logistic curve to the data using Sympy's nonlinsolve
# FIXME: objective_func = ...
objective_func = sum((logistic_func.subs(x, xi) - yi)**2 for xi, yi in zip(x_data, y_data))
gradient = sp.Matrix([sp.diff(objective_func, a), sp.diff(objective_func, b)])

# Initialize parameters (you can experiment with different initial values)
a_init, b_init = 1, 0

# Solve the nonlinear system using Newton's method
solution = sp.nsolve(gradient, [a, b], [a_init, b_init])
a_opt, b_opt = float(solution[0]), float(solution[1])  # Convert Sympy Floats to Python floats

# Plot the data and the fitted logistic curve
plt.scatter(x_data, y_data, label='Data')
x_fit = np.linspace(-5, 5, num=100)
y_fit = 1 / (1 + np.exp(-(a_opt * x_fit + b_opt)))  # Now this line should work without errors
plt.plot(x_fit, y_fit, label='Fitted Logistic Curve', color='red')
plt.legend()
plt.show()
```

### Multi-Layer Perceptron (MLP) with Finite Differentiation

In this exercise, we will implement a simple Multi-Layer Perceptron (MLP) from scratch using only NumPy and train it using gradient descent with finite differences.

A Multi-Layer Perceptron is a feedforward artificial neural network that consists of:
1. **Input layer**: receives the input features
2. **Hidden layer(s)**: performs non-linear transformations using activation functions
3. **Output layer**: produces the final prediction

**Architecture of our MLP:**
- Input layer: 2 neurons (for 2D input)
- Hidden layer: 4 neurons with sigmoid activation
- Output layer: 1 neuron with sigmoid activation (for binary classification)

**Forward pass equations:**
- Hidden layer: $h = \sigma(W_1 \cdot x + b_1)$ where $\sigma(z) = \frac{1}{1 + e^{-z}}$
- Output layer: $\hat{y} = \sigma(W_2 \cdot h + b_2)$

**Loss function (Binary Cross-Entropy):**
$$L = -\frac{1}{N} \sum_{i=1}^{N} \left[ y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right]$$

**Objective:**
Complete the code to:
1. Implement the forward pass
2. Compute the loss function
3. Calculate the numerical gradient using finite differences
4. Train the MLP using gradient descent

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Generate a simple random dataset
np.random.seed(42)
x_data = np.linspace(-5, 5, num=20)
y_data = 1 / (1 + np.exp(-x_data)) + np.random.normal(scale=0.1, size=len(x_data))

# Define the logistic function
x = sp.symbols('x')
a, b = sp.symbols('a, b')
# FIXME: logistic_func = ...
logistic_func = 1 / (1 + sp.exp(-(a * x + b)))

# Fit the logistic curve to the data using Sympy's nonlinsolve
# FIXME: objective_func = ...
objective_func = sum((logistic_func.subs(x, xi) - yi)**2 for xi, yi in zip(x_data, y_data))
gradient = sp.Matrix([sp.diff(objective_func, a), sp.diff(objective_func, b)])
gradient_func = sp.lambdify([a, b], gradient, 'numpy')

# Initialize parameters (you can experiment with different initial values)
a_init, b_init = 1, 0

# Perform Stochastic Gradient Descent (SGD)
learning_rate = 0.01
num_iterations = 1000
a_val, b_val = a_init, b_init

for iteration in range(num_iterations):
    # Compute symbolic gradient
    grad = gradient_func(a_val, b_val)
    grad_a, grad_b = float(grad[0][0]), float(grad[1][0])
    
    # Update parameters using gradient descent
    a_val -= learning_rate * grad_a
    b_val -= learning_rate * grad_b

a_opt, b_opt = a_val, b_val

# Plot the data and the fitted logistic curve
plt.scatter(x_data, y_data, label='Data')
x_fit = np.linspace(-5, 5, num=100)
y_fit = 1 / (1 + np.exp(-(a_opt * x_fit + b_opt)))  # Now this line should work without errors
plt.plot(x_fit, y_fit, label='Fitted Logistic Curve', color='red')
plt.legend()
plt.show()
```
